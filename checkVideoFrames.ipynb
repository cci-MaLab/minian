```python
import cv2
import os.path
```

```python
dpath="/N/project/Cortical_Calcium_Image/Miniscope data/10.2022_Fourth_group/AA024_D1/2022_10_08/12_35_50/Miniscope_2"
# videoPath="/N/project/Cortical_Calcium_Image/Miniscope data/12.2022_Sixth_group/AA033_D1/2022_12_04/10_42_18/Miniscope_2/0.avi"
# videoPath=videoPath.replace("\\","/").replace('"','').replace("'","").strip()
pathDir=os.listdir(dpath)
one_hour_frames=0
for video in pathDir:
    if video.endswith('.avi'):
        videoPath = dpath +'/' +video
        videoCapture=cv2.VideoCapture(videoPath)
        fps=videoCapture.get(cv2.CAP_PROP_FPS)
        frames=videoCapture.get(cv2.CAP_PROP_FRAME_COUNT)
        one_hour_frames=one_hour_frames+frames
        print(video)
        print("Number of frames:"+str(fps))
        print("Total frames:"+str(frames))
        print("Total video length:"+"{0:.2f}".format(frames/fps)+"seconds")
actual_fps=one_hour_frames/(60*60)
print("---------------------------------------------")
print("Total frames of one hour video:"+str(one_hour_frames))
print("Actual fps:"+str(actual_fps))
```
